```python
import dlt
```

```python
lookuptable_rules = {
    "rule1" : "show_id is not null"
}
```

```python
@dlt.table(
    name = "gold_netflix_directors"
)

@dlt.expect_all_or_drop(lookuptable_rules)
def myfunc():
    df = spark.readStream\
        .format('delta')\
    .load('abfss://silver@netflixstorageaccount.dfs.core.windows.net/netflix_directors')
    return df
```

```python
from pyspark.sql.functions import *
```

```python
@dlt.view

def gold_trns_netflixtitles():
    df = spark.readStream.table("LIVE.gold_stg_netflixtitles")
    df = df.withColumn("newflag", lit(1))
    return df
```
